```python
import pandas as pd
EDF_math=pd.read_csv('C:\\Users\\manvi\\OneDrive\\Documents\\Information Visualisation\\math-achievement-lea-sy2015-16.csv',sep=',')
EDF_math[:5]
```

```python
#Write and explain a function for processing a single column of “blurred” metrics into usable numeric values.
import re
def imputation(df,column_name): 
    col=df[column_name].astype(str)
    eval_col=[]
    for i in col:
        if  re.search(r" ", i):
            eval_col.append(np.nan)
        elif re.search('PS', i): 
            eval_col.append(np.nan)
        elif re.search(r'nan',i):
             eval_col.append(np.nan)
        elif re.search(r'\.',i):
             eval_col.append(np.nan)               
        elif re.search(r"LT.*", i):
            eval_col.append(((int(i.split('LT')[1])-1)+0)/2)
        elif re.search(r"LE.*", i):
            eval_col.append(((int(i.split('LE')[1]))+0)/2)
        elif re.search(r"GT.*", i):
            eval_col.append(((int(i.split('GT')[1])+1)+100)/2)
        elif re.search(r"GE.*", i):
            eval_col.append(((int(i.split('GE')[1]))+100)/2)
        elif re.search(r'-', i):
            eval_col.append((int(i.split('-')[0])+int(i.split('-')[1]))/2)         
        else:
            eval_col.append(int(i))
    df['imputed_col']=eval_col
    df['imputed_col'] = df['imputed_col'].fillna((df['imputed_col'].mean()))
    return(df)
```

```python
#Use it to process and then visualize the distribution of a performance metric of your choice.
# Visualising the distribution in terms of the total percentages

EDF_math_imputed=imputation(EDF_math,'ALL_MTH00PCTPROF_1516')
x=EDF_math_imputed['imputed_col']
num_bins = 20
n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
plt.ylabel('Number of students')
plt.xlabel('Bins for percentages')
plt.title('Distribution of percentage proficiency for Math ')
plt.show()
```

```python
#How much money is this?
fiscal_data=pd.read_csv('C:\\Users\\manvi\\OneDrive\\Documents\\Information Visualisation\\Sdf16_1a\\Sdf16_1a.txt',sep="\t")
fiscal_data=fiscal_data.filter(['TFEDREV','STNAME','LEAID','TOTALREV','TOTALEXP','TSTREV','TLOCREV'], axis=1)
fiscal_data_rev = fiscal_data[(fiscal_data[['TFEDREV']] > 0).all(axis=1)]
Remaining_amount=fiscal_data_rev['TFEDREV'].sum()*0.15

print("The total amount of Federal Revenue across all states is:",fiscal_data_rev['TFEDREV'].sum())
print("15% budget cut amounts to: ",fiscal_data_rev['TFEDREV'].sum()*0.15)
print("After cutting 15% of the revenue the amount remains:", (fiscal_data_rev['TFEDREV'].sum()-fiscal_data_rev['TFEDREV'].sum()*0.15))
```

```python
fiscal_data_rev = fiscal_data_rev.assign(SURPLUS = ((fiscal_data_rev['TOTALREV']) - fiscal_data_rev['TOTALEXP']))
fiscal_data_rev.sort_values(by='SURPLUS',axis=0, ascending=False)
fiscal_data_rev=fiscal_data_rev[fiscal_data_rev['SURPLUS']>=fiscal_data_rev['TFEDREV']]
c=0
leaid=[]
revenue=[]
for i, j in fiscal_data_rev.iterrows(): 
    if c<=Remaining_amount:
        leaid.append(j['LEAID'])
        revenue.append(j['TFEDREV'])
        c=c+j['TFEDREV']
```

```python
funding_cut = pd.DataFrame(
    {'LEAID': leaid,
     'CUT': revenue
    })
```
